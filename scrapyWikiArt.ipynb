```python
import json

data = []
with open('scrapWikiArt/data/canvas_metadata.json', 'r', encoding='utf-8') as f:
    for line in f:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from line: {line.strip()}\nError: {e}")
            continue

# If it's a list (which it will be now)
if isinstance(data, list):
    print(json.dumps(data[:3], indent=4))  # Show only first 3 items
# This else block might not be necessary after the fix, but kept for robustness
elif isinstance(data, dict):
    for key in list(data)[:3]:  # Show only first 3 keys
        print(f"{key}: {data[key]}")
```
